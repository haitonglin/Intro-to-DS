load WorldCupMatches.csv and WorldCupPlayers.csv to calculate all the matches that have goal in the first minute, and output the following dataframe.

![](ex5.png)

```python
import numpy as np
import pandas as pd
def myfunction():
    # YOUR CODE HERE
    dm=pd.read_csv('WorldCupMatches.csv')
    dp=pd.read_csv('WorldCupPlayers.csv')  
    dm.drop_duplicates(inplace=True)
    dm.dropna(inplace=True)
    dp.drop_duplicates(inplace=True)    #为什么不能合在一起写，一定要分开写？
    
    data=pd.merge(dm,dp,on='MatchID')
    
    data['first goal time']=data['Event'].str.extract('G(\d+)\'.*',expand=False).fillna(0).astype(int)
    
    data1=data.sort_values(by='first goal time',ascending=True)
    data1=data1[data1['first goal time']==1].set_index('MatchID')
    return data1
myfunction()
```

```python
import numpy as np
import pandas as pd
res=myfunction()
assert res.shape==(12,7)
assert res['Home Team Goals'].sum()==37
assert res['Away Team Goals'].sum()==20
```
