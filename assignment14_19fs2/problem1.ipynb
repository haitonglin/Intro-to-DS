# Problem 2

Draw the Maurer rose curve (https://en.wikipedia.org/wiki/Maurer_rose) when $n=6, d=71$.

Hint: $\frac{\pi}{180}\approx0.0174533$, polar coordinate system to Cartesian coordinate system (https://en.wikipedia.org/wiki/Polar_coordinate_system)

![](assignment14.png)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import math

def myplot():
    n = 6
    d = 71
    # YOUR CODE HERE
    X=[]
    Y=[]
    
    for a in range(361):
        
        k=a*d*math.pi/180
        r=math.sin(n*k)
        x=r*math.cos(k)
        y=r*math.sin(k)
        
        X.append(x)
        Y.append(y)
        
    ax=plt.axes()
    ax.plot(X,Y)
    ax.set(xlim=(-1.5,1.5),ylim=(-1.00,1.00),title='Maurer rose')
```

```python
myplot()
```
